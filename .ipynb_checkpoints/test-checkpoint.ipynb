```python
import math
import numpy as np
import warnings
warnings.filterwarnings('ignore', category=np.ComplexWarning)
from numpy import dtype, linalg
from scipy.linalg import expm
import matplotlib.pyplot as plt

import scipy

# Those are some functions I will use later to verify various results
def verify_orthogonality(eigenvecs):
    count = 0
    for i in range(eigenvecs.shape[0]):
        for j in range(eigenvecs.shape[1]):
            if i != j:  # Otherwise must be 1
                prod = np.vdot(eigenvecs[:, i], eigenvecs[:, j])
                if (abs(prod) > math.pow(10, -7)): count += 1

    if count == 0:
        print('Everything is ok with orthogonality')
    else:
        print('There is a problem with orthogonality')


# Check for unit length eigenvectors
def verify_normalization(eigenvecs):
    count = 0
    for i in range(eigenvecs.shape[0]):
        if abs(linalg.norm(eigenvecs[:, i]) - 1) > math.pow(10, -7): count += 1

    if count == 0:
        print('Everything is ok with normalization')
    else:
        print('There is a problem with normalization')


# Check if eigenvalue problem is satisfied
def verify_eigenvalue_problem(A, eigenvecs, eigenvals):
    count = 0
    for i in range(len(A)):
        calculation = np.matmul(A, eigenvecs[:, i]) - eigenvals[i] * eigenvecs[:, i]
        indeces = np.where(abs(calculation) > math.pow(10, -7))
        if not indeces:  # Check if the array is empty
            count += 1
    if count == 0:
        print('Eigenvalue problem is satisfied')
    else:
        print('There is a problem with Eigenvalue problem')


def run_checks(Hamiltonian):
    print('Some checks for the eigenvalue problem\n' + '-'*40)
    # d-> eigenvalues, Z- >eigenvectors
    d, Z = linalg.eigh(Hamiltonian)

    # Run some checks
    verify_orthogonality(Z)
    verify_normalization(Z)
    verify_eigenvalue_problem(Hamiltonian, Z, d)
    return d, Z


#https://www.kite.com/python/answers/how-to-check-for-duplicates-in-a-list-in-python
def check_degeneracy(eigenvals):
    print('-'*40+'\nDegeneracy:')
    if((len(eigenvals) == len(set(eigenvals))) ==True): print('First way:Negative')
    else: print('First way:Positive')

def check_degeneracy_second_way():
  #Case 1
  if chiA + chiD != 0:
    #Case 1.1
    if chiA + chiD > 0:
      #Case 1.1.1
      if( (omegaA-omegaD)/N >=-chiA and (omegaA-omegaD)/N <= chiD ): 
        print('Second way:Positive')
        return '1.1.1'
      #Case 1.1.2
      else: 
        print('Second way:Negative')
        return '1.1.2'
    #Case 1.2
    if chiA + chiD < 0:
      #Case 1.2.1
      if( (omegaA-omegaD)/N <=-chiA and (omegaA-omegaD)/N >= chiD ): 
        print('Second way:Positive')
        return '1.2.1'
      #Case 1.2.2
      else: 
        print('Second way:Negative')
        return '1.2.2'

  #Case 2
  else:
    #Case 2.1
    if (omegaD-omegaA-chiA*N >0 or omegaD-omegaA-chiA*N <0): 
      print('Second way:Negative')
      return '2.1'
    #Case 2.2
    else: 
      print('Second way:Energy is constant,Positive')
      return '2.2'

# To ensure that the first column corresponds to the case with 0 bosons at the donor

def check_probs(array,case):
    print('-' * 40)
    if (abs(np.sum(array) - 1) < math.pow(10, -7)): print('Sum of probs ok with ' + case)
    else:
        print('Probelm with the sum of probs with ' + case)
        print(np.sum(array))

N=12
omegaA = 2
omegaD = -3
chiA = 0.5
chiD = -0.5
coupling_lambda = 0.001

# ----------- Step 1: Construct the Hamiltonian matrix of the problem -----------

# Step 1a) Complete problem
H = np.zeros((N + 1, N + 1), dtype=float)

# Pointer m points to rows and n to columns. It is constructed based on the basis of the uncoupling problem
for m in range(N + 1):
    for n in range(N + 1):
        # First term from interaction
        if m == n + 1:
            H[m][n] = -coupling_lambda * np.sqrt((n + 1) * (N - n))
        # Second term from interaction
        if m == n - 1:
            H[m][n] = -coupling_lambda * np.sqrt(n * (N - n + 1))
        # Term coming from the two independent Hamiltonians
        if m == n:
            H[m][n] = omegaD * m + 0.5 * chiD * m ** 2 + omegaA * (N - m) + 0.5 * chiA * (N - m) ** 2

# Step 1b) Uncoupled case
H_uncoupling = np.zeros((N + 1, N + 1), dtype=float)
for m in range(N + 1):
    for n in range(N + 1):
        # Term coming from the two independent Hamiltonians
        if m == n:  H_uncoupling[m][n] = omegaD * m + 0.5 * chiD * m ** 2 + omegaA * (N - m) + 0.5 * chiA * (N - m) ** 2

# ----------- Step 2: Solve the eigenvalue problem -----------
eigenvalues, eigenvectors = run_checks(H)
#Diagonal matrix. Easy to compute eigenvalues
eigenvalues_uncoupling, eigenvectors_uncoupling = run_checks(H_uncoupling)

#Check if degeneracy
check_degeneracy(eigenvalues_uncoupling)
check_degeneracy_second_way()

#Make an indicative plot
Ns = np.linspace(0, N, num=N + 1, dtype=int)
plt.scatter(Ns, eigenvalues, label='Complete problem')
plt.scatter(Ns, eigenvalues_uncoupling, label='Uncoupled case')
plt.xlabel('i-th eigenvalue')
plt.ylabel('Energy')
plt.legend()
plt.title('Eigenvalues')
plt.show()

#Our theory applies only to the NON-DEGENERATE spectrum. In that cases we have to fix the eigenvectors in order
#for the ith eigenvalue to correspond to the nth eigenvector and vice versa and ensure that the first column corresponds 
#to the case with 0 bosons at the donor
def fix_eigenvectors(result,eigvals,eigvecs):
  print('-'*40+'\nNow we have to relate the ith eigenvalue with the number of bosons')
  if result == '1.1.2': 
    print('\n Either increasing either decreasing. We will use the values at n=0,n=N to deduce what is happening')
    if(omegaD*N + 0.5*chiD*N**2 - omegaA*N + 0.5*chiA*N**2 > 0):  print('Energy is increasing, no need for extra corrections')
    else: eigvals,eigvecs = np.flip(eigvals),np.fliplr(eigvecs)

  if result == '1.2.2':
    print('\n Either increasing either decreasing.We will use the values at n=0,n=N to deduce what is happening')
    if(omegaD*N + 0.5*chiD*N**2 - omegaA*N + 0.5*chiA*N**2 > 0):  print('Energy is increasing, no need for extra corrections')
    else: eigvals,eigvecs = np.flip(eigvals),np.fliplr(eigvecs)

  if (result =='2.1' and omegaD-omegaA-chiA*N >0):print('Energy is increasing, no need for extra corrections')

  if (result =='2.1' and omegaD-omegaA-chiA*N <0):
    print('Energy is decreasing, must flip both eigenvectors and eigenvalues')
    eigvals,eigvecs = np.flip(eigvals),np.fliplr(eigvecs)
  
  plt.plot(Ns,eigvals)
  plt.xlabel('Number of bosons in donor')
  plt.ylabel('Energy of the uncoupling system')
  plt.show()
  return eigvals,eigvecs 

eigenvalues,eigenvectors = fix_eigenvectors(check_degeneracy_second_way(),eigenvalues,eigenvectors)

initial_state = np.zeros(N + 1, dtype=float)
for n in range(N + 1):
    initial_state[n] = np.exp(n-(N/2))
initial_state = initial_state / np.linalg.norm(initial_state)

P = np.zeros(N+1)
for n in range(N + 1):
    P[n] = (np.abs( np.vdot(eigenvectors[:, n], initial_state) ) ) ** 2
check_probs(P,'Pertubation')

plt.plot([i for i in range(len(P))], P)
plt.plot([i for i in range(len(initial_state))], initial_state)
plt.show()

# ----------- Step 3: Compute average number of bosons -----------

# Create an orthonormal basis of HS
A = scipy.linalg.orth(H_uncoupling)

coeff_c = np.zeros((N+1, N+1))
for i in range(N+1):
    for j in range(N):
        coeff_c[i,j] = np.vdot(eigenvectors[:,j].T, eigenvectors_uncoupling[:,i])

coeff_b = np.zeros(N+1)
for k in range(N+1):
    coeff_b[k] = np.vdot(eigenvectors_uncoupling[:,k].T, initial_state)

coeff_a = np.zeros(N+1)
t = np.linspace(0,10,10, dtype=int)
avg_N2 = []
for time in t:
    sum_n = 0
    for j in range(N+1):
        sum_a = 0
        for i in range(N+1):
            sum_a += coeff_b[i]*coeff_c[i,j]*np.exp(-complex(0,1)*eigenvalues_uncoupling[i]*time)
        coeff_a[j] = sum_a
        sum_n += j*np.abs(coeff_a[j])**2
    print(sum_n)
    avg_N2.append(sum_n)

plt.plot(t, avg_N2)
plt.xlabel('Time')
plt.ylabel(r'<N$_D(t)$>')
plt.show()

```

```python
import math
import numpy as np
import warnings
warnings.filterwarnings('ignore', category=np.ComplexWarning)
from numpy import dtype, linalg
from scipy.linalg import expm
import matplotlib.pyplot as plt
import time
import scipy
import pandas as pd
from google.colab import files
import shutil

def construct_Hamiltonians(chiA,chiD):
  H,H_uncoupling = np.zeros((N + 1, N + 1), dtype=float),np.zeros((N + 1, N + 1), dtype=float)
  
  # Step 1a) Complete problem
  # i bosons at the donor
  for i in range(N + 1):
    for j in range(N + 1):
      # First term from interaction
      if i == j - 1: H[i][j] = -coupling_lambda * np.sqrt((i + 1) * (N - i))
      # Second term from interaction
      if i == j + 1:H[i][j] = -coupling_lambda * np.sqrt(i * (N - i + 1))
        # Term coming from the two independent Hamiltonians
      if i == j: H[i][j] = omegaD * i + 0.5 * chiD * i ** 2 + omegaA * (N - i) + 0.5 * chiA * (N - i) ** 2

  # Step 1b) Uncoupling problem
  np.fill_diagonal(H_uncoupling,np.diagonal(H))

  return H,H_uncoupling

# Those are some functions I will use later to verify various results
def verify_orthogonality(eigenvecs):
    count = 0
    for i in range(eigenvecs.shape[0]):
        for j in range(eigenvecs.shape[1]):
            if i != j:  # Otherwise must be 1
                prod = np.vdot(eigenvecs[:, i], eigenvecs[:, j])
                if (abs(prod) > math.pow(10, -7)): count += 1

    if count == 0:
        print('Everything is ok with orthogonality')
    else:
        print('There is a problem with orthogonality')


# Check for unit length eigenvectors
def verify_normalization(eigenvecs):
    count = 0
    for i in range(eigenvecs.shape[0]):
        if abs(linalg.norm(eigenvecs[:, i]) - 1) > math.pow(10, -7): count += 1

    if count == 0:
        print('Everything is ok with normalization')
    else:
        print('There is a problem with normalization')


# Check if eigenvalue problem is satisfied
def verify_eigenvalue_problem(A, eigenvecs, eigenvals):
    count = 0
    for i in range(len(A)):
        calculation = np.matmul(A, eigenvecs[:, i]) - eigenvals[i] * eigenvecs[:, i]
        indeces = np.where(abs(calculation) > math.pow(10, -7))
        if not indeces:  # Check if the array is empty
            count += 1
    if count == 0:
        print('Eigenvalue problem is satisfied')
    else:
        print('There is a problem with Eigenvalue problem')


def run_checks(Hamiltonian):
    print('Some checks for the eigenvalue problem\n' + '-'*40)
    # d-> eigenvalues, Z- >eigenvectors
    d, Z = linalg.eigh(Hamiltonian)

    # Run some checks
    verify_orthogonality(Z)
    verify_normalization(Z)
    verify_eigenvalue_problem(Hamiltonian, Z, d)
    return d, Z


#https://www.kite.com/python/answers/how-to-check-for-duplicates-in-a-list-in-python
def check_degeneracy(eigenvals):
    print('-'*40+'\nDegeneracy:')
    if((len(eigenvals) == len(set(eigenvals))) ==True): print('First way:Negative')
    else: print('First way:Positive')

def check_degeneracy_second_way(chiA,chiD):
  #Case 1
  if chiA + chiD != 0:
    #Case 1.1
    if chiA + chiD > 0:
      #Case 1.1.1
      if( (omegaA-omegaD)/N >=-chiA and (omegaA-omegaD)/N <= chiD ): 
        print('Second way:Positive')
        return '1.1.1'
      #Case 1.1.2
      else: 
        print('Second way:Negative')
        return '1.1.2'
    #Case 1.2
    if chiA + chiD < 0:
      #Case 1.2.1
      if( (omegaA-omegaD)/N <=-chiA and (omegaA-omegaD)/N >= chiD ): 
        print('Second way:Positive')
        return '1.2.1'
      #Case 1.2.2
      else: 
        print('Second way:Negative')
        return '1.2.2'

  #Case 2
  else:
    #Case 2.1
    if (omegaD-omegaA-chiA*N >0 or omegaD-omegaA-chiA*N <0): 
      print('Second way:Negative')
      return '2.1'
    #Case 2.2
    else: 
      print('Second way:Energy is constant,Positive')
      return '2.2'

# To ensure that the first column corresponds to the case with 0 bosons at the donor

def check_probs(array,case):
    print('-' * 40)
    if (abs(np.sum(array) - 1) < math.pow(10, -7)): print('Sum of probs ok with ' + case)
    else:
        print('Probelm with the sum of probs with ' + case)
        print(np.sum(array))

def An(n,chiA,chiD):
    f1 = omegaA + 0.5 * chiA * (2*N - 2*n - 1) - omegaA - 0.5 * chiD * (2 * n + 1)
    return -np.sqrt((n + 1) * (N-n)) / f1

def Bn(n,chiA,chiD):
    f2 = -omegaA - 0.5 * chiA * (2*N - 2*n + 1) + omegaD + 0.5 * chiD * (2*n - 1)
    return -np.sqrt(n * (N- n + 1)) / f2

def Cn(n,chiA,chiD):
    f3 = 2*(omegaA-omegaD-chiD*(n+1) + chiA * (N -n - 1) ) 
    return -An(n,chiA,chiD)*np.sqrt((n+2) * (N-n-1)) / f3    

def Dn(n,chiA,chiD):
    f4 = 2*(-omegaA+omegaD+chiD*(n-1)-chiA*(N-n+1)) 
    return -Bn(n,chiA,chiD)*np.sqrt((n-1) * (N- n + 2)) / f4

def Fn(n,chiA,chiD):
  x = 1/np.sqrt(1+coupling_lambda**2*(An(n,chiA,chiD)**2 + Bn(n,chiA,chiD)**2) + 
                coupling_lambda**4*(Cn(n,chiA,chiD)**2 + Dn(n,chiA,chiD)**2))
  return x

def Gram_Schmidt(A):
  Q = np.zeros(( len(A),len(A) ),dtype=float)
  Q[:,0] =A[:,0]/np.linalg.norm(A[:,0])
  for q_indice in range(1,len(A)):
    crops = 0
    for indice in range(q_indice):
      crops += np.vdot(Q[:,indice], A[:,q_indice])*Q[:,indice]
    Q[:,q_indice] = A[:,q_indice] - crops
    Q[:,q_indice] = Q[:,q_indice]/np.linalg.norm(Q[:,q_indice])
  return Q

def bosons_at_donor_analytical(max_t,eigvecs,eigvals):
  coeff_c = np.zeros(N+1,dtype=float)
  for i in range(N+1): 
    coeff_c[i] = np.vdot(eigvecs[:,i], initial_state)

  j_vectors = np.identity(N+1)

  coeff_b = np.zeros(eigvecs.shape)
  for j in range(N+1):
    for i in range(N+1):
      coeff_b[j,i] = np.vdot(j_vectors[:, j], eigvecs[:, i])

def sum_js(t,n,k,i,eigvals,eigvecs,eigvecs_tilde,C_eigvecs):
  sum_j_return = 0
  for j in range(N+1):
    sum_j_return += np.exp(1j*(eigenvalues[n]-eigenvalues[j])*t)*eigvecs[k,j]*np.vdot(eigvecs[:,j].T,eigvecs_tilde[:,i])
  return sum_j_return

def bosons_at_donor_analytical(max_t,eigvecs,eigvals,initial_state):
  coeff_c = np.zeros(N+1,dtype=float)
  for i in range(N+1): 
    coeff_c[i] = np.vdot(eigvecs[:,i], initial_state)

  j_vectors = np.identity(N+1)

  coeff_b = np.zeros(eigvecs.shape)
  for j in range(N+1):
    for i in range(N+1):
      coeff_b[j,i] = np.vdot(j_vectors[:, j], eigvecs[:, i])

  # %%
  avg_N = np.zeros(max_t+1,dtype=float)
  counter = 0
  t = 0

  while True and t <= max_t:
    sum_m = 0
    for m in range(N+1):
      sum_i = 0
      for i in range(N+1):
        sum_i += coeff_c[i]*coeff_b[m,i]*np.exp(-1j*eigvals[i]*t)

      sum_k = 0
      for k in range(N+1):
        sum_k += coeff_c[k].conj()*coeff_b[m,k].conj()*np.exp(1j*eigvals[k]*t)*sum_i
    
      sum_m += sum_k*m
    #print("\rt={}".format(t), end = "")
    avg_N[t] = sum_m
    #if avg_N[counter] < 0.1: break
    #counter += 1
    t += 1

  return avg_N

def sum_js(t,n,k,i,eigvals,eigvecs,eigvecs_tilde,C_eigvecs):
  sum_j_return = 0
  for j in range(N+1):
    sum_j_return += np.exp(1j*(eigenvalues[n]-eigenvalues[j])*t)*eigvecs[k,j]*np.vdot(eigvecs[:,j].T,eigvecs_tilde[:,i])
  return sum_j_return

def bosons_at_donor_perturbation(max_t,eigvecs,eigvecs_tilde,eigvals):
  C_eigvecs = np.zeros(N+1,dtype=float)
  avg_ND = np.zeros(max_t+1,dtype=float)
  for l in range(N+1):C_eigvecs[l] = np.vdot(eigvecs_tilde[:,l].T, initial_state)
  
  t=0
  while t<=max_t:
    print("\rt={}".format(t), end = "")
    sum_k=0
    for k in range(N+1):
      sum_i=0
      for i in range(N+1):
        sum_m = 0
        for m in range(N+1):
          sum_n =0
          for n in range(N+1):
            sum_j = sum_js(t,n,k,i,eigvals=eigenvalues,eigvecs=eigenvectors,eigvecs_tilde=eigenvectors_pert,C_eigvecs=C_eigvecs)
            sum_n += np.conjugate(eigvecs[k,n])*np.vdot(eigvecs_tilde[:,m].T,eigvecs[:,n])*sum_j

          sum_m += np.conjugate(C_eigvecs[m])*sum_n
        sum_i += C_eigvecs[i]*sum_m
      sum_k += k*sum_i

    avg_ND[t] = sum_k
    t += 1
    

  return avg_ND


def write_data(data,name_of_file):
  df = pd.DataFrame(data = data)	
  np.savetxt(name_of_file, df)



def read_2D_data(name_of_file):
	X,Y = [],[]
	for line in open(name_of_file, 'r'):
  		lines = [i for i in line.split()]
  		X.append(float(lines[0]))
  		Y.append(float(lines[1]))
	return X,Y

def read_1D_data(name_of_file):
	data = []
	for line in open(name_of_file, 'r'):
  		lines = [i for i in line.split()]
  		data.append(float(lines[0]))
	return data

"""
#Our theory applies only to the NON-DEGENERATE spectrum. In that cases we have to fix the eigenvectors in order
#for the ith eigenvalue to correspond to the nth eigenvector and vice versa and ensure that the first column corresponds 
#to the case with 0 bosons at the donor
def fix_eigenvectors(result,eigvals,eigvecs):
  print('-'*40+'\nNow we have to relate the ith eigenvalue with the number of bosons')
  if result == '1.1.2': 
    print('\n Either increasing either decreasing. We will use the values at n=0,n=N to deduce what is happening')
    if(omegaD*N + 0.5*chiD*N**2 - omegaA*N + 0.5*chiA*N**2 > 0):  print('Energy is increasing, no need for extra corrections')
    else: eigvals,eigvecs = np.flip(eigvals),np.fliplr(eigvecs)

  if result == '1.2.2':
    print('\n Either increasing either decreasing.We will use the values at n=0,n=N to deduce what is happening')
    if(omegaD*N + 0.5*chiD*N**2 - omegaA*N + 0.5*chiA*N**2 > 0):  print('Energy is increasing, no need for extra corrections')
    else: eigvals,eigvecs = np.flip(eigvals),np.fliplr(eigvecs)

  if (result =='2.1' and omegaD-omegaA-chiA*N >0):print('Energy is increasing, no need for extra corrections')

  if (result =='2.1' and omegaD-omegaA-chiA*N <0):
    print('Energy is decreasing, must flip both eigenvectors and eigenvalues')
    eigvals,eigvecs = np.flip(eigvals),np.fliplr(eigvecs)
  
  plt.plot(Ns,eigvals)
  plt.xlabel('Number of bosons in donor')
  plt.ylabel('Energy of the uncoupling system')
  plt.show()
  return eigvals,eigvecs 
"""

# ----------- Step 0: Set the parameters of the problem -----------
start_time = time.time()

N = 12
omegaA,omegaD = 3,-3
#chiA,chiD = -0.5,0.5
coupling_lambda = 0.001

xA = np.linspace(-4,4,100)
xD = np.linspace(-4,4,100)

def execute(chiA,chiD):
  # ----------- Step 1: Construct the Hamiltonian matrices of the problem -----------
  H,H_uncoupling = construct_Hamiltonians(chiA,chiD)

  # ----------- Step 2: Solve the eigenvalue problem -----------
  eigenvalues, eigenvectors = linalg.eigh(H)
  #Diagonal matrix. Easy to compute eigenvalues
  eigenvalues_uncoupling, eigenvectors_uncoupling = linalg.eigh(H_uncoupling)
  #Check if degeneracy
  #check_degeneracy(eigenvalues_uncoupling)
  #check_degeneracy_second_way(chiA,chiD)

  #----------- Step 3: Initial state ----------- 
  initial_state = np.zeros(N+1,dtype=float)
  for n in range(N+1): initial_state[n] = np.exp(-(N-n)**2)
  initial_state = initial_state / np.linalg.norm(initial_state)
  
  #----------- Step 4: Time evolution of the wavefuction with perturbation -----------

  #----------- Step 4a: Analytical solution ----------- 


  #----------- Step 4b: Perturbation approach ----------- 
  """
  eigenvectors_pert, eigenvalues_pert = np.zeros((N + 1, N + 1), dtype=float), np.zeros((N + 1, N + 1), dtype=float)
  #Create and normalize the eigenvectors with pertubation theory
  eigenvectors_pert[:, 0] = (eigenvectors_uncoupling[:,0] + 
                            coupling_lambda*An(0,chiA,chiD)*eigenvectors_uncoupling[:,1] + 
                            coupling_lambda**2*Cn(0,chiA,chiD)*eigenvectors_uncoupling[:,2])

  eigenvectors_pert[:, 0] = eigenvectors_pert[:,0]/np.sqrt(1+coupling_lambda**2*An(0,chiA,chiD)**2 + coupling_lambda**4*Cn(0,chiA,chiD)**2)

  eigenvectors_pert[:, 1] = (eigenvectors_uncoupling[:,1] +
                            coupling_lambda*An(1,chiA,chiD) * eigenvectors_uncoupling[:,2] + 
                            coupling_lambda*Bn(1,chiA,chiD)*eigenvectors_uncoupling[:,0] +
                            coupling_lambda**2*Cn(1,chiA,chiD)*eigenvectors_uncoupling[:, 3])
  eigenvectors_pert[:, 1] = eigenvectors_pert[:,1]*Fn(1,chiA,chiD)

  eigenvectors_pert[:, N] = (eigenvectors_uncoupling[:,N] + 
                            coupling_lambda*Bn(N,chiA,chiD)*eigenvectors_uncoupling[:,N-1] +
                            coupling_lambda**2*Dn(N,chiA,chiD) * eigenvectors_uncoupling[:,N-2] )
  eigenvectors_pert[:, N] = eigenvectors_pert[:,N]/np.sqrt(1+coupling_lambda**2*Bn(N,chiA,chiD)**2 + coupling_lambda**4*Dn(N,chiA,chiD)**2)

  eigenvectors_pert[:, N-1] = (eigenvectors_uncoupling[:,N-1] + 
                             coupling_lambda*An(N-1,chiA,chiD)*eigenvectors_uncoupling[:,N] + coupling_lambda*Bn(N-1,chiA,chiD)*eigenvectors_uncoupling[:,N-2] + 
                             coupling_lambda**2*Dn(N-1,chiA,chiD)*eigenvectors_uncoupling[:,N-3] )
  eigenvectors_pert[:, N-1] = eigenvectors_pert[:, N-1]*Fn(N-1,chiA,chiD)


  for n in range(2, N-1):
    eigenvectors_pert[:,n]=(eigenvectors_uncoupling[:,n]+
                            coupling_lambda*An(n,chiA,chiD)*eigenvectors_uncoupling[:,n+1] + 
                            coupling_lambda*Bn(n,chiA,chiD)*eigenvectors_uncoupling[:,n-1] +
                            coupling_lambda**2*Cn(n,chiA,chiD)*eigenvectors_uncoupling[:,n+2] + 
                            coupling_lambda**2*Dn(n,chiA,chiD)*eigenvectors_uncoupling[:,n-2])
    eigenvectors_pert[:,n] = Fn(n,chiA,chiD)*eigenvectors_pert[:, n]
    eigenvalues_pert[n] = (eigenvalues_uncoupling[n] - 
                          coupling_lambda**2*(An(n,chiA,chiD)*np.sqrt((n+1)*(N-n)) + 
                                              Bn(n,chiA,chiD)*np.sqrt(n*(N-n+1)) ) )


  # In order to make them orthogonal, we have to follow Gram-Schmidt normalization
  eigenvectors_pert = Gram_Schmidt(eigenvectors_pert)
  #verify_eigenvalue_problem(H,eigenvectors_pert,eigenvalues_pert)
  #verify_normalization(eigenvectors_pert)
  #verify_orthogonality(eigenvectors_pert)


  #----------- Step 5: Plot results ----------- 
  #plt.figure(figsize=(8,6))
  """
  t_max = 2000
  t_span = np.linspace(0,t_max,t_max+1)
  avg_ND_analytical = bosons_at_donor_analytical(max_t=t_max,eigvecs=eigenvectors,eigvals=eigenvalues,initial_state=initial_state)
  """
  #avg_ND_perturbation = bosons_at_donor_perturbation(max_t=t_max,eigvecs=eigenvectors,eigvecs_tilde=eigenvectors_pert,eigvals=eigenvalues)
  plt.scatter(t_span,avg_ND_analytical,c='b',s=70,label='Analytical solution')
  #plt.scatter(t_span,avg_ND_perturbation,c='r',label='Perturbation solution')
  plt.plot(t_span,avg_ND_analytical,c='0')
  #plt.plot(t_span,avg_ND_perturbation,c='0')
  plt.xlabel('Time')
  plt.ylabel(r'$<N_D(t)>$')
  plt.legend()
  """
  #print(f'\nExecution time ={round(time.time() - start_time,4)} seconds')
  title_file = f'ND_analytical,λ={coupling_lambda},χΑ={chiA},χD={chiD}.txt'
  write_data(avg_ND_analytical,name_of_file =title_file)
  shutil.move("/content/" + title_file, "/content/new_data")
  #files.download(title_file)
  #write_data(avg_ND_perturbation,name_of_file =f'ND_perturbation,λ={coupling_lambda}.txt')


!mkdir new_data

count_it=0
for chiA in xA:
  for chiD in xD:
    count_it +=1
    print("\rCombination {} out of {}: (chiA,chiD) = ({},{})".format(count_it,len(xA)*len(xD),round(chiA,4),round(chiD,4)), end = " ")
    execute(chiA,chiD)

!zip -r "/content/new_data_zipped.zip" "/content/new_data" 
```
